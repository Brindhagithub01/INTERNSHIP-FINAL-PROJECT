```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define the columns you want to use as features
columns_to_use = ['num_refill_req_l3m', 'transport_issue_l1y', 'Competitor_in_mkt', 'retail_shop_num', 'distributor_num',
                    'flood_impacted', 'flood_proof', 'dist_from_hub', 'workers_num', 'wh_est_year','storage_issue_reported_l3m',
                    'temp_reg_mach', 'wh_breakdown_l3m', 'govt_check_l3m','product_wg_ton']

# Extract features and target variable from your training and testing datasets
X_train = train1[columns_to_use]
y_train = train1['product_wg_ton']
X_test = test1[columns_to_use]
y_test = test1['product_wg_ton']

# Random Forest Regression
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Decision Tree Regression
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
y_pred_dt = dt_regressor.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

# Gradient Boosting Regression
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_regressor.fit(X_train, y_train)
y_pred_gb = gb_regressor.predict(X_test)
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

# AdaBoost Regression with Decision Tree as base learner
adaboost_regressor = AdaBoostRegressor(estimator=DecisionTreeRegressor(random_state=42), n_estimators=50, random_state=42)
adaboost_regressor.fit(X_train, y_train)
y_pred_adaboost = adaboost_regressor.predict(X_test)
mse_adaboost = mean_squared_error(y_test, y_pred_adaboost)
r2_adaboost = r2_score(y_test, y_pred_adaboost)

# Print results
print("Random Forest Regression Results:")
print(f"Mean Squared Error: {mse_rf}")
print(f"R-squared (R2) Score: {r2_rf}")
print("Accuracy:",rf_regressor.score(X_test, y_test)*100)

print("\nDecision Tree Regression Results:")
print(f"Mean Squared Error: {mse_dt}")
print(f"R-squared (R2) Score: {r2_dt}")
print("Accuracy:",dt_regressor.score(X_test, y_test)*100)

print("\nGradient Boosting Regression Results:")
print(f"Mean Squared Error: {mse_gb}")
print(f"R-squared (R2) Score: {r2_gb}")
print("Accuracy:",gb_regressor.score(X_test, y_test)*100)

print("\nAdaBoost Regression with Decision Tree Base Learner Results:")
print(f"Mean Squared Error: {mse_adaboost}")
print(f"R-squared (R2) Score: {r2_adaboost}")
print("Accuracy:",adaboost_regressor.score(X_test, y_test)*100)
```
